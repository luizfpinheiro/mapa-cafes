```python
import folium
```

```python
import pandas as pd
```

```python
mapa = folium.Map(location=[-15.8155646,-48.0072168], tiles="Stamen Toner", zoom_start=11.8)
```

```python
mapa
```

```python
df = pd.read_csv('data-cafe.csv', encoding='utf-8')
```

```python
df.columns = ['Cafeteria', 'Mês', 'Lat', 'Long']
```

```python
df
```

```python
points = []
for i in range(df.shape[0]):
    points.append([df['Lat'][i], df['Long'][i]])
    
    folium.Marker([df['Lat'][i], df['Long'][i]], 
                  tooltip=df['Cafeteria'][i],
                  icon=folium.Icon(color='yellow', icon='coffee', prefix='fa')).add_to(mapa)
    
```

```python
mapa
```

```python
folium.PolyLine(points, color="black", opacity=0.5).add_to(mapa)
```
