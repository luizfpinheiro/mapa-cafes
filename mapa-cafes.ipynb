```python
import branca
import folium
from folium.features import CustomIcon
```

```python
import pandas as pd
```

```python
mapa = folium.Map(location=[-15.80,-48.0011], tiles='Stamen Terrain', zoom_start=11.8)
# tiles: cartodbpositron - cartodbdark_matter - Stamen Terrain - Stamen Toner
mapa
```

```python
df = pd.read_csv('data-cafe.csv', encoding='utf-8')
```

```python
df.columns = ['Cafeteria', 'Mês', 'Lat', 'Long']
```

```python
df.to_json(orient='records', force_ascii=False)
```

```python

points = []

for i in range(df.shape[0]):
    points.append([df['Lat'][i], df['Long'][i]])
    
    html = """<div class="mapouter"><div class="gmap_canvas"><iframe width="350" height="250" id="gmap_canvas" """ + """src="https://maps.google.com/maps?q={}&t=&z=13&ie=UTF8&iwloc=&output=embed" """.format(df['Cafeteria'][i]) + """frameborder="0" scrolling="no" marginheight="0" marginwidth="0"></iframe></div></div>"""
    
    
    iframe = branca.element.IFrame(html=html, width=600, height=270)
    popup = folium.Popup(iframe, max_width=350)
    
    folium.Marker([df['Lat'][i], df['Long'][i]], 
                  tooltip=df['Cafeteria'][i],
                  popup=popup,
                  icon=folium.Icon(color='black', icon_color='lightgray', icon='coffee', prefix='fa')).add_to(mapa)
```
